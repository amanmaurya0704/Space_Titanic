```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("C:/Users/amanm/Downloads/Machine Learning/dataset/classification/fruits.csv")
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
sc = StandardScaler()
X1 = sc.fit_transform(X)
model = KNeighborsClassifier()
model.fit(X,y)
```

```python
sample = [[2.5,70]]
sample= sc.transform(sample)
model.predict(sample)
```

```python
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv("C:/Users/amanm/Downloads/Machine Learning/dataset/classification/fruits.csv")
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values


model = KNeighborsClassifier()
sc = StandardScaler()

pipe = Pipeline([("sc",sc),("m",model)])
pipe.fit(X,y)
```

```python
sample = [[2.5,70]]
pipe.predict(sample)
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
X= iris.data
y = iris.target
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X,y)
```

```python
model.predict([[2.6,1.2,4.8,.4]])
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plot_tree(model,fontsize=8,feature_names=['SL','SW',''])
plt.show()
```
